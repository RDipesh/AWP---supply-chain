▶**Total sale in every location?**

```python
sns.barplot(x=data["Location"],y=data["Number of products sold"])
plt.xlabel("Location" )
plt.ylabel("Total Sales")
plt.title("Total Sales in Every Location")
plt.xticks(rotation=45)
linewidth=(0)
plt.show()
```

▶**What is the average shipping cost per location in the dataset?**

▶**How can we calculate the average shipping time for each shipping method?**

▶**What is the distribution of price by Location**

```python
 price_dis_loc=data.groupby("Location")["Price"].sum()
```
